```python
import json
import os

# Input JSON file path (replace this with your actual filename)
input_file = "momentum.json"

# Output directory for generated files
output_dir = "filtered_outputs"
os.makedirs(output_dir, exist_ok=True)

# Load your original JSON data
with open(input_file, "r") as f:
    data = json.load(f)

# Create a dictionary to collect data for each filter
filters_dict = {}

# Loop through each month and its filters
for month, filters in data.items():
    for filter_name, stocks in filters.items():
        if filter_name not in filters_dict:
            filters_dict[filter_name] = {}
        filters_dict[filter_name][month] = stocks

# Write each filter’s data into a separate JSON file
for filter_name, month_data in filters_dict.items():
    output_filename = f"Momentum_{filter_name}.json"
    output_path = os.path.join(output_dir, output_filename)

    with open(output_path, "w") as outfile:
        json.dump(month_data, outfile, indent=4)

    print(f"✅ Saved: {output_filename}")

print("\nAll filter files created successfully!")
```
