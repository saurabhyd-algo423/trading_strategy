```python
import json
import pandas as pd
```

```python
value_df = pd.read_csv("2025-10-13 (1).csv")
momentum_df = pd.read_csv('u1_stock_list.csv')
```

```python
with open('names.json','r') as file:
    tickers = json.load(file)
```

```python
print(tickers)
```

```python
value_df['ROE_ROCE_greater_than_15'].unique()
```

```python
momentum_df.columns
```

```python
final_df = pd.DataFrame()
```

```python
tickers = [ticker+'.NS' for ticker in tickers]
```

```python
print(tickers)
```

```python
final_df['Portfolio Ticker'] =  tickers
```

```python
final_df
```

```python
# Assuming 'final_df' is the DataFrame where you want to add the column.
final_df['Value Satisfied'] = False # Initialize the column

# Set 'Ticker' as the index of value_df for fast lookups (optional but better)
value_df_indexed = value_df.set_index('Ticker')

for index, ticker in final_df['Portfolio Ticker'].items():
    # 1. Check if the ticker exists in the value_df index
    if ticker in value_df_indexed.index:
        # 2. Use .loc[] to safely look up the 'ROE_ROCE_greater_than_15' status
        status = value_df_indexed.loc[ticker, 'ROE_ROCE_greater_than_15']

        # NOTE: .loc[ticker] could return a Series (multiple rows) if Tickers aren't unique.
        # This code assumes a single unique Ticker match.
        if (status == True):
            final_df.loc[index, 'Value Satisfied'] = True
```

```python
final_df
```

```python
final_df['Value Satisfied'].value_counts()
```

```python
# Assuming 'final_df' is the DataFrame where you want to add the column.
final_df['Momentum Satisfied'] = False # Initialize the column

# Set 'Ticker' as the index of value_df for fast lookups (optional but better)
value_df_indexed = momentum_df.set_index('Stock')

for index, ticker in final_df['Portfolio Ticker'].items():
    # 1. Check if the ticker exists in the value_df index
    if ticker in value_df_indexed.index:
        final_df.loc[index, 'Momentum Satisfied'] = True
```

```python
final_df
```

```python
final_df['Momentum Satisfied'].value_counts()
```

```python
final_df['Value Momentum Satisfied'] = final_df['Value Satisfied'] | final_df['Momentum Satisfied']
final_df
```
